Importing **Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Importing the Dataset:**

```python
dataset =pd.read_csv("/content/weatherAUS[1].csv")

```

```python
dataset['Date']=pd.to_datetime(dataset['Date'])
dataset['year']=dataset['Date'].dt.year
dataset['month']=dataset['Date'].dt.month
dataset['day']=dataset['Date'].dt.day
```

```python
dataset.drop(['Date'],axis=1,inplace=True)
```

```python
dataset['RainToday'].replace({'No':0,'Yes':1},inplace=True)
dataset['RainTomorrow'].replace({'No':0,'Yes':1},inplace=True)
```

```python
dataset.tail()
```

**Checking and dealing with imbalance of data:**

```python
print('Yes RainTomorrow counts:',dataset[(dataset['RainTomorrow']==1)].RainTomorrow.count())
print('No RainTomorrow counts:',dataset[(dataset['RainTomorrow']==0)].RainTomorrow.count())
```

```python
from sklearn.utils import resample
no_rt=dataset[dataset['RainTomorrow']==0]
yes_rt=dataset[dataset['RainTomorrow']==1]
oversampled_yes_rt=resample(yes_rt,replace=True,n_samples=len(no_rt),random_state=137)
```

```python
fil_dataset=pd.concat([no_rt,oversampled_yes_rt])
```

```python
print('Yes RainTomorrow counts:',fil_dataset[(fil_dataset['RainTomorrow']==1)].RainTomorrow.count())
print('No RainTomorrow counts:',fil_dataset[(fil_dataset['RainTomorrow']==0)].RainTomorrow.count())
```

```python
print(fil_dataset.select_dtypes(include=['object']).columns)
```

```python
print(fil_dataset.isnull().sum())
```

**Dealing with null values of Categorical Variables:**

```python
list=['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm']
for item in list:
  fil_dataset[item]=fil_dataset[item].fillna(fil_dataset[item].mode()[0])
```

```python
print(fil_dataset.isnull().sum())
```

**Dealing with null values of Numerical variables:**

```python
fil_dataset[fil_dataset.columns[fil_dataset.isna().any()]].columns
```

```python
null_list=['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
       'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am',
       'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm',
       'Temp9am', 'Temp3pm', 'RainToday']
for k in null_list:
  fil_dataset[k]=fil_dataset[k].fillna(fil_dataset[k].mean())
```

```python
print(fil_dataset.isnull().sum())
```

**Encoding Categorical variables :**

```python
from sklearn.preprocessing import LabelEncoder
encoders={}
for j in list:
  encoders[j]=LabelEncoder()
  fil_dataset[j]=encoders[j].fit_transform(fil_dataset[j])
```

```python
print(fil_dataset.dtypes)
```

```python
fil_dataset.shape
```

```python
fil_dataset.head()
```

```python
x=fil_dataset.iloc[:,fil_dataset.columns!='RainTomorrow']
y=fil_dataset['RainTomorrow']
```

**Scaling the dataset:**

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
x=sc.fit_transform(x)
```

**Selecting K most relevant features:**

```python
from sklearn.feature_selection import chi2,SelectKBest
k_best=SelectKBest(score_func=chi2,k=15)
k_best.fit(x,y)
```

```python
x_new=k_best.transform(x)
```

**Splitting the dataset:**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_new,y,test_size=0.3,stratify=y,random_state=137,shuffle=True)
```

**Building and training the model:**

```python
import xgboost as xgb
xgb_param={'n_estimators':500,
           'max_depth':16}
model=xgb.XGBClassifier(**xgb_param)
model.fit(x_train,y_train)
```

**Evaluating the model:**

```python
from sklearn.metrics import classification_report
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cf=confusion_matrix(y_test,y_pred)
sns.heatmap(cf,annot=True)
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.show()
```
